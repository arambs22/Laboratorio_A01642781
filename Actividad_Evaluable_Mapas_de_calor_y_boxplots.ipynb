```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("clientes-centro-comercial.csv")
```

```python
sns.pairplot(data = df, hue = 'Genero')
plt.title("Clientes Centro Comercial")
```

```python
corr = df.corr()
sns.heatmap(data = corr, vmin = -1, vmax = 1, cmap = "YlOrRd", annot = True)
```

```python
fig = plt.figure(figsize = (9,5))
sns.boxplot(data = df, x = "Edad")
```

```python
fig = plt.figure(figsize = (9,5))
sns.boxplot(data = df, x = "Ingresos")
```

```python
fig = plt.figure(figsize = (9,5))
sns.boxplot(data = df, x = "Puntos")
```
